```r
# Load the datasets
sessions <- read.csv("data/sessions.csv")
players <- read.csv("data/players.csv")

sessions
players
```

```r
#1. Distribution of Played Hours
played_hours <- ggplot(players, aes(x = played_hours)) +
  geom_histogram(binwidth = 5, fill = "steelblue", color = "black") +
  labs(title = "Distribution of Total Played Hours",
       x = "Total Played Hours",
       y = "Number of Players") +
  theme(text = element_text(size = 20))

played_hours
```

```r
#2. Line chart comparing age with their mean amount of playing hours
age_vs_mean_hours <- players |>
    filter(age < 51) |>
    group_by(age) |>
    summarize(avg_played_hours = mean(played_hours, na.rm = TRUE), .groups = 'drop') |>
    ggplot(aes(x = age, y = avg_played_hours)) +
    geom_line(color = "dodgerblue", size = 1) +
    geom_point(color = "darkblue", size = 2) +
    labs(x = "Age",
         y = "Average Hours Played",
         title = "Trend of Average Hours Played Across Ages") +
    theme_minimal(base_size = 18)

age_vs_mean_hours
```

```r
#3. Boxplot of Played Hours by Experience Level
played_hours_by_experience <- ggplot(players, aes(x = experience, y = played_hours, fill = experience)) +
  geom_boxplot() +
  labs(title = "Played Hours by Experience Level",
       x = "Experience Level",
       y = "Total Played Hours") +
  theme(text = element_text(size = 20))

played_hours_by_experience
```

```r
#4. Stacked bar chart of total hours by experience level and gender
experience_vs_hours <- players |>
    group_by(experience, gender) |>
    summarize(total_hours = sum(played_hours, na.rm = TRUE), .groups = 'drop') |>
    ggplot(aes(x = experience, y = total_hours, fill = gender)) +
    geom_bar(stat = "identity") +
    labs(x = "Experience Level",
         y = "Total Hours Played",
         title = "Total Hours Played by Experience Level and Gender") +
    theme(text = element_text(size = 20)) +
    scale_fill_brewer(palette = 'Dark2')

experience_vs_hours

```

```r
#5. Bar Plot of Subscription Status
subscriptin_status <- ggplot(players, aes(x = subscribe)) +
  geom_bar(fill = "pink", color = "black") +
  labs(title = "Count of Subscription Status",
       x = "Subscription Status",
       y = "Number of Players") +
  theme(text = element_text(size = 20))

subscriptin_status
```

```r
#6. Stacked percentage bar chart for gender make-up within each experience level
gender_makeup <- players |>
    group_by(experience, gender) |>
    summarize(count = n(), .groups = 'drop') |>
    ggplot(aes(x = experience, y = count, fill = gender)) +
    geom_bar(stat = "identity", position = "fill") +
    labs(x = "Experience Level",
         y = "Proportion of Gender (%)",
         title = "Proportion of Gender by Experience Level") +
    theme(text = element_text(size = 20)) +
    scale_fill_brewer(palette = 'Set2')

gender_makeup
```

```r
#7. Faceted line chart showing mean hours by age across experience levels
players_bar <- players |>
    filter(age < 51) |>
    group_by(age, experience) |>
    summarize(avg_played_hours = mean(played_hours, na.rm = TRUE), .groups = 'drop') |>
    ggplot(aes(x = age, y = avg_played_hours, color = experience)) +
    geom_line(size = 1) +
    facet_wrap(~ experience, scales = "free_y") +
    labs(x = "Age",
         y = "Average Hours Played",
         color = "Experience Level",
         title = "Average Hours Played by Age and Experience Level") +
    theme(text = element_text(size = 20)) +
    scale_color_manual(values = c("dodgerblue", "darkorange", "darkorchid", "springgreen", "gold"))

players_bar
```
